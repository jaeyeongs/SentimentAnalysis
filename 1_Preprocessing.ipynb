### (2) 데이터 전처리

- 결측치 & 중복 제거

```python
df.isnull().sum() # NULL 값 유무 확인
```

```python
df['review'].nunique() # review 변수에 있는 중복내용을 제외한 리뷰
```

```python
df.drop_duplicates(subset=['review'], inplace=True) # review 열에서 중복인 내용이 있다면 중복 제거
print('리뷰의 수 :',len(df))
```

- \n(개행문자) 제거

```python
# \n(개행문자) -> ""(공백)
df['review'] = df['review'].apply(lambda x: x.replace('\n',''))
```

- 텍스트에 포함되어 있는 특수 문자 제거

```python
import re
def cleanText(readData):
 
    
    text = re.sub('[-=+,#/\?:;^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》❤]', '', readData)
    return text

df['review'] = df['review'].apply(cleanText)
```

- 의미없는 자음 & 모음으로 이루어진 단어 제거

```python
def cleanText2(readData):
 
    text = re.sub('[ㄱㄴㄷㄹㅁㅂㅅㅇㅈㅊㅋㅌㅍㅎㅃㅉㄸㄲㅆㅛㅕㅑㅐㅔㅗㅓㅏㅣㅜㅠㅡ]', '', readData)
    return text
df['review'] = df['review'].apply(cleanText2)
```

- 이모티콘 제거

```python
def cleanText3(readData):
    
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

    text = emoji_pattern.sub(r'', readData)
    return text

df['review'] = df['review'].apply(cleanText3)
```

- 맞춤법 검사

```python
# 맞춤법 검사기 hanspell 사용
from hanspell import spell_checker
```

```python
def spell_check(review):
    result = spell_checker.check(review)
    return result.checked
```

```python
df['review']=df['review'].astype('str')
```

```python
df['review_spell_check'] = df['review'].apply(spell_check)
```

```python
df # 맞춤범 검사 후 데이터 확인
```

```python
df = df.loc[:, ["category", "store", "id", "review_spell_check", "star"]]
df
```
